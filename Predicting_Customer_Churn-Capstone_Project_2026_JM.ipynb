## Decision Tree Modelling

Decision Tree Classifier for initial modelling

After establishing a baseline, a Decision Tree model is used as the first predictive model to capture patterns in customer behaviour.

I am using Decision Tree Classifier to fit a model tree_1 on the training data with all default settings. This will score the data using accuracy on the train and test data.

```python
# Initial Decision Tree with default settings
model = DecisionTreeClassifier(random_state=22)
model.fit(X_train, y_train)

print("Train accuracy:", model.score(X_train, y_train))
print("Test accuracy:", model.score(X_test, y_test))
```

#### Analysis:
The model achieves nearly perfect accuracy (99.8%) on training data but significantly lower accuracy (84.8%) on test data. This large gap of approximately 15% indicates overfitting - the model has memorized the training data rather than learning generalizable patterns. This suggests the tree has grown too deep with default settings, capturing noise instead of meaningful relationships.

#### Hyperparameter Optimization:

To address the overfitting issue, I will systematically optimize two key hyperparameters that control model complexity: max_depth (limiting tree growth) and ccp_alpha (cost complexity pruning). This approach helps find the optimal balance between model complexity and generalization performance.

```python
# Test different depths and pruning techniques
depths = range(1, 21)
alphas = [0, 0.0001, 0.001, 0.01, 0.1]

# Depth optimization
train_scores = []
test_scores = []

for d in depths:
    dt = DecisionTreeClassifier(max_depth=d, random_state=22)
    dt.fit(X_train, y_train)
    train_scores.append(dt.score(X_train, y_train))
    test_scores.append(dt.score(X_test, y_test))

# Pruning optimization
ccp_train_scores = []
ccp_test_scores = []

for alpha in alphas:
    tree = DecisionTreeClassifier(ccp_alpha=alpha, random_state=22)
    tree.fit(X_train, y_train)
    ccp_train_scores.append(tree.score(X_train, y_train))
    ccp_test_scores.append(tree.score(X_test, y_test))

# Plot both optimizations
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Depth plot
ax1.plot(depths, train_scores, '--o', label='Train Accuracy')
ax1.plot(depths, test_scores, '--o', label='Test Accuracy')
ax1.plot(np.argmax(test_scores)+1, max(test_scores), 'v', markersize=13, color='red', label='Optimal Depth')
ax1.set_xlabel('Tree Depth')
ax1.set_ylabel('Accuracy')
ax1.set_title('Depth Optimization')
ax1.legend()
ax1.grid(True)

# Pruning plot
ax2.plot(alphas, ccp_train_scores, '--o', label='Train Accuracy')
ax2.plot(alphas, ccp_test_scores, '--o', label='Test Accuracy')
ax2.set_xlabel('ccp_alpha (Pruning Strength)')
ax2.set_ylabel('Accuracy')
ax2.set_title('Pruning Optimization')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.savefig("images/decision_tree_optimization.png", dpi=300)
plt.show()

# Find optimal parameters
optimal_depth = depths[np.argmax(test_scores)]
optimal_alpha = alphas[np.argmax(ccp_test_scores)]

print(f"Optimal depth: {optimal_depth}")
print(f"Optimal alpha: {optimal_alpha}")
```

The optimization reveals that a maximum depth of 9 provides the best test accuracy, significantly reducing complexity from the default unlimited depth. The ccp_alpha of 0.0001 indicates minimal pruning is needed, suggesting the depth constraint is the primary factor in preventing overfitting. These optimized parameters should substantially reduce the training-test accuracy gap while maintaining strong predictive performance.

```python
# Apply optimized parameters
model = DecisionTreeClassifier(max_depth=9)
model.fit(X_train, y_train)

print("Train accuracy:", model.score(X_train, y_train))
print("Test accuracy:", model.score(X_test, y_test))
```

#### Analysis:
Using the optimized max_depth of 9 (as identified in the previous optimization step), the final decision tree model achieves:

Train Accuracy: 91.61%
Test Accuracy: 88.58%

Compared to the default model, this optimized decision tree shows:

- Reduced overfitting: The gap between training and test accuracy is now much smaller (around 3%), indicating the model has learned generalizable patterns rather than just memorizing the training data.
- Improved generalization: The test accuracy has increased from 84.8% to 88.58%, demonstrating better performance on unseen data.

While the training accuracy is still higher than the test accuracy, the smaller gap suggests the model has found a good balance between complexity and generalization, making it more suitable for real-world deployment.

Overall, the systematic optimization of the max_depth parameter has yielded a more robust and effective decision tree classifier.

## Clustering (K-mean)

While predictive models identify who is likely to churn, clustering helps explain why customers behave differently by grouping similar behavioural patterns.


Predicting that a customer will churn is only half the battle. To actually stop them, a business needs to know who they are and why they are leaving.
- The Goal: To segment the customer base into distinct "personas" (e.g., the "High-Spender at Risk" vs. the "Low-Engagement Browser").
- The Benefit: It allows for Targeted Action. You wouldn't send the same email to a loyal customer who is frustrated with support as you would to a new user who just hasn't logged in. Clustering provides the "why" behind the churn signal.
  
## Scaling the data


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# 1. Select features and DROP rows with missing values
X_cluster_data = df_encoded.drop('Churned', axis=1).dropna() # .dropna() fixes the error!

# 2. Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster_data)

# 3. Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=22, n_init=10)
# We use X_cluster_data.index to make sure the labels match the right rows
clusters = kmeans.fit_predict(X_scaled)

# 4. Attach back to a copy of the dataframe so we don't lose data
df_clustered = df.loc[X_cluster_data.index].copy()
df_clustered['Cluster'] = clusters

print(df_clustered['Cluster'].value_counts())
```

### Cluster Profiling

```python
cluster_profile = df_clustered.groupby('Cluster').mean(numeric_only=True)
print(cluster_profile)
```

The clustering results group customers into segments based on behaviour and engagement patterns. By analysing the average characteristics of each cluster, we can identify distinct customer profiles such as high-value customers, low-engagement users, and high-risk churn groups.

These clusters can be used to design targeted retention strategies, as different customer segments require different interventions.
